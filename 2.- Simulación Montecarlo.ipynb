# <center> **Simulación Montecarlo**

El método de Montecarlo es una estrategia de resolución de problemas que involucra el uso de la estadística. Sea $P$ la probabilidad de que ocurra determinado evento $A$, entonces podemos simular aleatoriamente este suceso y obtener $P$ hallando el cociente entre el número de veces que ocurrió nuestro suceso y el número de simulaciones totales $N$, de la siguiente manera

\begin{equation}
P = \frac{Número \ de \ veces \ que \ ocurre \ el \ evento \ A}{Número \ total \ de\  eventos N}
\end{equation}



## **Estimación de $\pi$ mediante simulación montecarlo** 

Vamos a aplicar la simulación montecarlopara obtener una aproximación de $\pi$. Pi ($\pi$) es una constante matemática que nos indica la relación entre la longitud de una circunferencia y su diámetro. Si denotamos por $C$ la longitud de una circunferencia y por $d$ su diámetro, sabemos que $C = d* \pi$. La longitud de una circunferencia de diametro $1$ vale $\pi$.

Antes de realizar la simulación vamos a considerar el plano cartesiono para valores positivos, es decir, vamos a emplear $1/4$ del área total del círculo, por lo que el valor estimado de $\pi$ debe ser multiplicado por cuatro. Vamos a denotar $M$ como el número de puntos que caen dentro de la región de la circunferencia, mientras que $N$ el total de valores simulados.

\begin{equation}
\pi = 4 \times \frac{M}{N}
\end{equation}

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
N = 1000
M = 0
```

```python
X_in=[]
Y_in=[]
X_out=[]
Y_out=[]
for i in range(N):
  x=random.random()
  y=random.random()
  if(x**2+y**2 <= 1):
    M+=1
    X_in.append(x)
    Y_in.append(y)
  else:
    X_out.append(x)
    Y_out.append(y)
```

```python
pi = 4*M/N
error = abs((pi - np.pi) / pi) * 100
error
```

```python
# circunferencia
x = np.linspace(0,1)
y=[math.sqrt(1-val**2) for val in x]
```

```python
plt.figure(figsize=(8,8))
plt.plot (x , y, color='blue' , linewidth='2')
plt.scatter(X_in, Y_in, color='orange' , marker='.')
plt.scatter(X_out, Y_out, color='cyan', marker='.')
plt.grid (which='major')
plt.title ('Estimación de $\pi$')
plt.show()
```

```python

def mc_pi_aprox(N=1000):

  X_in=[]
  Y_in=[]
  X_out=[]
  Y_out=[]
  M = 0
  for i in range(N):
    x=random.random()
    y=random.random()
    if(x**2+y**2 <= 1):
      M+=1
      X_in.append(x)
      Y_in.append(y)
    else:
      X_out.append(x)
      Y_out.append(y)

  # calculo de pi y error    
  pi = 4*M/N
  error = abs((pi - np.pi) / pi) * 100
  # Circunferencia
  x_c = np.linspace(0,1)
  y_c =[np.sqrt(1-val**2) for val in x_c]
  # gráfico
  plt.figure(figsize=(8,8))  # tamaño de la figura
  plt.plot (x_c , y_c, color='red' , linewidth='2')
  plt.plot(X_in, Y_in, 'b.')
  plt.plot(X_out, Y_out, 'r.')
  plt.plot(0, 0, label='$\hat\pi$ = {:4.8f}\nerror = {:4.2f}%'.format(pi,error), alpha=0)
  plt.axis('square')
  plt.legend(frameon=True, framealpha=0.9, fontsize=16)
```

```python
mc_pi_aprox(N=1000)
```

```python

def mc_pi_aprox(N=1000):

  X_in=[]
  Y_in=[]
  X_out=[]
  Y_out=[]
  M = 0
  for i in range(N):
    x = np.random.uniform(-1, 1, size=1)
    y = np.random.uniform(-1, 1, size=1)
    if(x**2+y**2 <= 1):
      M+=1
      X_in.append(x)
      Y_in.append(y)
    else:
      X_out.append(x)
      Y_out.append(y)

  # calculo de pi y error    
  pi = 4*M/N
  error = abs((pi - np.pi) / pi) * 100
  # Circunferencia
  x_c = np.linspace(-1,1, 1000)
  y_sup =[np.sqrt(1-val**2) for val in x_c]
  y_inf =[-np.sqrt(1-val**2) for val in x_c]
  # gráfico
  plt.figure(figsize=(8,8))  # tamaño de la figura
  plt.plot (x_c , y_sup, color='red' , linewidth='2')
  plt.plot (x_c , y_inf, color='red' , linewidth='2')
  plt.plot(X_in, Y_in, 'b.')
  plt.plot(X_out, Y_out, 'r.')
  plt.plot(0, 0, label='$\hat\pi$ = {:4.8f}\nerror = {:4.2f}%'.format(pi,error), alpha=0)
  plt.axis('square')
  plt.legend(frameon=True, framealpha=0.9, fontsize=16)
```

```python
mc_pi_aprox(N=1000)
```
